```python
import matplotlib.pyplot as plt
import random

# Sample text data
text_data = """
Nature's beauty is a timeless marvel, with lush green forests, sparkling rivers, and majestic mountains. The vibrant colors of flowers in bloom paint a picturesque scene, while the gentle whispers of the wind create a symphony of tranquility. Life is a journey filled with moments of joy, love, and adventure.
"""

# Convert text to lowercase
text_data = text_data.lower()

# Tokenize the text
words = text_data.split()

# Define a list of stopwords
stopwords_list = ["this", "is", "a", "for", "to", "you", "can", "and", "in", "by", "the", "or", "are", "of"]

# Remove stopwords
filtered_words = [word for word in words if word.isalpha() and word not in stopwords_list]

# Create a dictionary to store word frequencies
word_freq = {}

# Count word frequencies
for word in filtered_words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

# Sort the words by frequency
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Display the word cloud without overlapping using a grid-based approach
plt.figure(figsize=(10,6))
grid_size = 15
grid = [[False] * grid_size for _ in range(grid_size)]

for word, freq in sorted_word_freq:
    placed = False
    while not placed:
        x = random.randint(0, grid_size - 2)
        y = random.randint(0, grid_size - 2)

        if not grid[x][y]:
            grid[x][y] = True
            placed = True

    color = (random.random(), random.random(), random.random())  # Random RGB color
    plt.text(x / grid_size, y / grid_size, word, fontsize=freq * 15, ha='center', va='center', color=color)

plt.axis("off")
plt.show()
```
